```python
total_nulos = df_sales.isnull().sum().sort_values(ascending = False)
total_nulos
```

```python
porcentaje_nulos = (df_sales.isnull().sum()/df_sales.isnull().count()).sort_values(ascending=False)
porcentaje_nulos
```

```python
#Convertir 'porcentaje_nulos' en DataFrame
missing_data = pd.concat([total_nulos, porcentaje_nulos], axis=1, keys=['total_nulos', 'porcentaje_nulos'])
missing_data
```
