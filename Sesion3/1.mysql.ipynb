```python
import sqlalchemy as db
from sqlalchemy import text
import pandas as pd
```

```python
#engine = db.create_engine("mysql://user:pass@ip:port/database.sql")
engine = db.create_engine("mysql://root:root@172.16.5.4:3310/retail_db")
```

```python
conn = engine.connect()
```

```python
customer_df = pd.read_sql("SELECT * FROM customers", con=conn)
customer_df.head()
```

```python
order_df = pd.read_sql("SELECT * FROM orders", con=conn)
order_df.head()
```

```python
order_item_df = pd.read_sql("SELECT * FROM order_items", con=conn)
order_item_df.head()
```

```python
product_df = pd.read_sql("SELECT * FROM products", con=conn)
product_df.head()
```

```python
categories_df = pd.read_sql("SELECT * FROM categories", con=conn)
categories_df.head()
```

```python
department_df = pd.read_sql("SELECT * FROM departments", con=conn)
department_df.head()
```

```python
# Promedio de ingresos por día

df_promedio = order_df.merge(order_item_df, left_on="order_id", right_on="order_item_order_id", how="left")
df_promedio.head()
```

```python
df_promedio_final = df_promedio[['order_date', 'order_item_subtotal']].groupby('order_date').sum()['order_item_subtotal']/df_promedio[['order_date', 'order_item_order_id']]\
                    .drop_duplicates().groupby('order_date').count()['order_item_order_id']
df_promedio_final
```

```python
# Promedio de ingreso por mes
df_promedio['month'] = df_promedio['order_date'].astype(str).str[5:7] #la parte 5:7 obtiene del string 2013-07-25, que indica que desde el indice 5 al 6, que al final queda 07
df_promedio['month'] = df_promedio['month'].astype('int32')
df_promedio.head()
```

```python
df_promedio_final = df_promedio[['month', 'order_item_subtotal']].groupby('month').sum()['order_item_subtotal']/df_promedio[['month', 'order_item_order_id']]\
                    .drop_duplicates().groupby('month').count()['order_item_order_id']
df_promedio_final
```

```python
df_join  = order_df.merge(order_item_df, left_on='order_id', right_on='order_item_order_id')\
                    .merge(product_df, left_on='order_item_product_id', right_on='product_id')\
                    .merge(categories_df, left_on='product_category_id', right_on='category_id')\
                    .merge(department_df, left_on='category_department_id', right_on='department_id')

df_join
```

```python
df_promedio = df_join[['department_name','order_item_subtotal']].groupby('department_name').sum()
df_promedio.head()
```

```python
# Promedio de venta por departamentos
df_promedio = df_join[['department_name', 'order_item_subtotal']].groupby('department_name').sum()['order_item_subtotal']/df_join[['department_name', 'order_item_order_id']]\
                    .drop_duplicates().groupby('department_name').count()['order_item_order_id']
df_promedio.head()
```
